```python
import requests as r
import pandas as pd
from bs4 import BeautifulSoup
```

```python
link="https://en.wikipedia.org/wiki/Bhagat_Singh"
res=r.get(link)
soup=BeautifulSoup(res.text,'html.parser')
title=soup.find('span',class_='mw-page-title-main').text
print(title)
```

```python
data=str()
for i in soup.find_all('p'):
    data=data+i.text
    data=data+"\n"
data=data.replace('\u20b9','')
print(data)
fname=title+".txt"
fd=open(fname,'w')
fd.write(data)
fd.close()
```
